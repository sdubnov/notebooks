# Introduction to Digital Audio

How to read audio into the computer?

```python
import scipy.io.wavfile as wav
import matplotlib.pyplot as plt

x = wav.read("flute.wav")
print x
sampling_freq = x[0]
audio = x[1]

plt.plot(audio)
plt.ylabel("Amplitude") 
plt.xlabel("Time (samples)") 
# set the title
plt.title("Flute Sample")
```

Let's take a closer look at the samples themselves:

```python
plt.plot(audio[20000:21000])
```

This resembles a sinusoid. So we can create a sound from a sinwave formula!

But what is the frequency?
We have approximately 11.25 periods in 1000 samples. So the period is ...

```python
period = 1000./11.25
freq = 1/period

synth = 15000*sin(2*pi*freq*arange(60000.))
plt.plot(synth[20000:21000])

wav.write('out.wav',sampling_freq,synth.astype('int16'))
```

Let's just quickly correct the fade in..

```python
env = ones(shape(synth))
env[0:30000] = linspace(0,1,30000)
synthe = env*synth
plot(synthe)

wav.write('out.wav',sampling_freq,synthe.astype('int16'))
```
