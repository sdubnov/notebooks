## Reading and Writing MIDI Files

A very different manner of representing music is by means of so called MIDI files.
MIDI stands for Musical Instruments Digital Interface, and is in fact just a communication
standard between synthesizer.

It is important to realize that MIDI represents musical performance actions, and not the sound itself. It is actually a rather complicated collection of data... 

First we use a program that translates midi file to csv.
There are other ways to read midi files, also native to python, but I like this method since it allows treating music scores / midi using tools such as pandas.

./midicsv ../jacques.mid > ../jacques.csv 

```python
import numpy as np
from pandas import DataFrame, read_csv

def csv2song(notes_csv):
    midi_events = notes_csv[' Header']
    noteon = np.where(midi_events == ' Note_on_c')[0]
    noteoff = np.where(midi_events == ' Note_off_c')[0]
    noteson = np.array([notes_csv.loc[i][[1,4,5]] for i in noteon])
    
    keytckon = noteson[:,0].astype(int)
    keynum = noteson[:,1].astype(int)
    keyvel = noteson[:,2].astype(int)
    
    #tuple of 4 (start_tick, note, length ,velocity)
    song = []
    
    if len(noteoff) == 0:
        onset = {}
        hold = []
        for i in range(0,len(keynum)):
            if keynum[i] in hold:
                song.append((onset[keynum[i]][0], keynum[i], keytckon[i]-onset[keynum[i]][0], onset[keynum[i]][1]))
                hold.remove(keynum[i])
                del onset[keynum[i]]
            else:
                hold.append(keynum[i])
                onset[keynum[i]] = (keytckon[i],keyvel[i])
    else:
        notesoff = np.array([notes_csv.loc[i][[1,4,5]] for i in noteoff])
        keytckoff = notesoff[:,0].astype(int)
        for i in range(0,len(keynum)):
            song.append((keytckon[i], keynum[i], keytckoff[i]-keytckon[i], keyvel[i]))
            
    return song
```

Lets try now to recreate the file using more traditional music notation.
Here is how it works: 
60 is middle C, let's assume 128 is a quarter note, and let's not worry about velocity.

```python
from __future__ import division
from pyknon.genmidi import Midi
from pyknon.music import NoteSeq, Note

notes1 = NoteSeq("D4 E F# D8 R D4 E F# D8 R F#4 G A2 F#4 G A2")

midi = Midi(number_tracks=1, tempo=90)

midi.seq_notes(notes1, track=0)

midi.write("./jacques-out.mid")
```

Go to the appropriate folder and listen to the two midi files. Of course they sound the same - this is how jacques.midi was created on the first place :)
